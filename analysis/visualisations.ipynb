```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('../cleaned/twitter_with_emotion_toxicity.csv')

# Step 1: Select only columns that start with "emo_"
emotion_cols = [col for col in df.columns if col.startswith('emo_')]

# Step 2: Calculate correlation matrix between emotions
emotion_corr = df[emotion_cols].corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    emotion_corr,
    annot=True,             # Display correlation numbers
    cmap='coolwarm',         # Color palette
    fmt=".2f",               # 2 decimal places
    linewidths=0.5,          # Lines between squares
    cbar_kws={'label': 'Correlation'}
)
plt.title(' Heatmap of Emotions Co-occurrence', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
